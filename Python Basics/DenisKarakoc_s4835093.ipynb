
1.2.2 Principal Component Analysis, more commonly known as PCA, can be cleverly used to better visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.

**Principal Component Analysis is a transformation used to visualize data or pre-process it for data mining by reducing a dataset's dimensionality. The first principal component is the direction in which the data varies the most. The second principal component is orthogonal to the first and thus captures leftover variance. The third principal component is orthogonal to the second, and so on. By projecting the data onto the principal components, one can get rid of superflous dimensions and noise in the data while preserving the core variability. 
Since variance and the mean of a dataset are not necessarily its best representation, PCA does not always work. Moreover, by projecting onto principal components, one also loses variance in other directions, which might not always be noise, but its essence.** 

To apply PCA we must first mean center the data. mean centering means that the mean value for an attribute is subtracted from that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

**By mean centering data it is normalized. If attributes vary greatly in magnitude of scale, then the greater scale attribute will dominate the variance, leading to a skewed representation of the data.**

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X**, must first be mean centered. This can be easily done by calculating a row vector **μ** containing the mean values of each attribute and then subtracting it from **X** (**Y** = **X** - **1μ**, where **1** is a column vector of size **N** x 1, with N indicating the number of observations in the original data).

Then, the Singular Value Decomposition, or SVD, of **Y**, the mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the numpy.linalg.svd() function.

Using SVD on **Y** yields a series of vectors which can be used to project the data onto specific Principal Components (PCs). The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,1]** would yield the projection of the data onto the first PC.

> d. (1 point) Alternatively, one could use EigenValue Decomposition, EVD, instead of SVD. What are the similarities and differences between SVD and EVD? Can both be applied in each case?

**EVD and SVD are very similar, with SVD being a more generalized version of the underlying process. The main idea is to decompose a matrix into 3 transformations: a change of basis to the core of the data (the eigenvectors in EVD, a set of orthonormal vectors in SVD), a scaling of the data (the eigenvalues and singular values represtively), and finally a change of basis to the output dimensions (eigenvectors' inverse in EVD and a reverse mapping rotation in SVD). Since Eigenvalues exist only for square matrices (and not even all of them, e.g. rotation), EVD is much more limited than SVD. But EVD is faster when it can be applied, because the mean centering is included in its process (since the decomposition is performed on the covariance matrix).**

PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values.

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 gray scale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\end{eqnarray*}
where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary
data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary,
we will transform the data to be binary when calculating these two measures of
similarity by setting
$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$

\begin{eqnarray*}
\textrm{Cosine}(\alpha{x},{y}) & = & \textrm{Cosine}({x},{y}) \\
\end{eqnarray*}
This does not hold. The cosine similarity measures the angle between two vectors. Hence, if a vector is stretched, it should not affect its direction. However, if it is stretched by a negative scalar, it will flip its direction, changing the angle with it.

\begin{eqnarray*}
\textrm{Cosine}(\beta + {x},{y}) & = & \textrm{Cosine}({x},{y}) \\
\end{eqnarray*}
This obviously does not hold. Adding a constant to a vector changes its direction and the angle between it and another vector.

\begin{eqnarray*}
\textrm{ExtendedJaccard}(\alpha{x},{y}) & = & \textrm{ExtendedJaccard}({x},{y}) \\
\end{eqnarray*}
The extended Jaccard takes into account the angle between vectors (like the cosine similarity) but also how long they are (their norm). Scaling the vector can change both direction and length, thus both angle and norm. That is why the equation above is false.

\begin{eqnarray*}
\textrm{ExtendedJaccard}(\beta + {x},{y}) & = & \textrm{ExtendedJaccard}({x},{y}) \\
\end{eqnarray*}
The workings of the extended Jaccard are explained above. Translating the vector does change its angle and its norm. The above equation still does not hold, because of the angle difference.

\begin{eqnarray*}
\textrm{Correlation}(\alpha{x},{y}) & = & \textrm{Correlation}({x},{y}) \\
\end{eqnarray*}
This does not hold. Stretching a vector increases its variance. This, in turn, disproportionately effects the correlation. If A and B both increase in steps of 1, and then suddenly B' increases in steps of 2, those two correlations will not be the same. 
Multiplying by a negative number flips a vector around, making a positive correlation negative, and vice versa.


\begin{eqnarray*}
\textrm{Correlation}(\beta + {x},{y}) & = & \textrm{Correlation}({x},{y})
\end{eqnarray*}
Adding a constant to a vector changes its direction, which obviously needs to effect the correlation (parts that previously increased in a vector could now be even decreasing). The rate of increase/decrease is sure to be altered.
